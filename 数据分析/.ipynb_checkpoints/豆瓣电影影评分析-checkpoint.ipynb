## 好评率计算

```python
# 去除 star 为“暂未评价”的数据
df_star = df.query("star != '暂未评价'")
score = {
    '力荐':5,
    '推荐':4,
    '还行':3,
    '较差':2,
    '很差':1
}
df_star['star'] = df_star["star"].map(score)
```

```python
df_star.groupby("film").mean()
```
